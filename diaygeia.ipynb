```python
import requests
import time
import pandas as pd


keywords = [
    "ανακαίνιση σχολείου",
    "ανακαίνιση σχολείων",
    "συντήρηση σχολείου",
    "συντήρηση σχολείων",
    "συντήρηση σχολικών",
    "επισκευή σχολείου",
    "επισκευή σχολείων",
    "ενεργειακή αναβάθμιση σχολ"
]

all_decisions = []
base_url = 'https://diavgeia.gov.gr/luminapi/api/search'

for kw in keywords:
    print(f"--- Συλλογή για: '{kw}' ---")
    page = 0
    size = 100

    while True:
        params = {
            'q': f'subject:"{kw}"',
            'size': size,
            'page': page
        }

        response = requests.get(base_url, params=params)

        if response.status_code != 200:
            print(f"Παράκαμψη {kw} στη σελίδα {page} λόγω σφάλματος {response.status_code}")
            break

        data = response.json()
        decisions = data.get("decisions", [])

        if not decisions:
            break

        all_decisions.extend(decisions)
        print(f"Σελίδα {page} | Τρέχον σύνολο εγγραφών: {len(all_decisions)}")

        page += 1
        time.sleep(0.5)

print(f"\nΤέλος συλλογής! Μαζεύτηκαν συνολικά {len(all_decisions)} εγγραφές (μαζί με διπλότυπα).")
```

```python
import pandas as pd

# 1. Μετατροπή της φρέσκιας λίστας σε DataFrame
df_raw = pd.DataFrame(all_decisions)
print(f"Αρχικές εγγραφές στη λίστα: {len(df_raw)}")

# 2. Εξαγωγή του καθαρού ονόματος του Δήμου
df_raw['foreas_name'] = df_raw['organization'].apply(
    lambda x: x.get('label', 'Άγνωστος Φορέας') if isinstance(x, dict) else 'Χωρίς Φορέα'
)

# 3. Αφαίρεση διπλοτύπων με βάση τον μοναδικό ΑΔΑ (Deduplication)
df_unique = df_raw.drop_duplicates(subset=['ada']).copy()
print(f"Μοναδικές εγγραφές μετά την αφαίρεση διπλοτύπων: {len(df_unique)}")
```

```python
# ============================================================
# Νεες διορθωσεις σε σχεση με την προηγουμενη εκδοση:
#  1) Το "Ω" σε καποια PDF ειναι στην πραγματικοτητα το συμβολο OHM
#     (U+2126, απο τα ηλεκτρολογικα) και οχι το κανονικο ελληνικο Ω
#     (U+03A9) -- οπτικα ιδια, διαφορετικος χαρακτηρας. Διορθωνεται
#     οπως και το ∆ -> Δ.
#  2) Πολλα εντυπα γραφουν τις επικεφαλιδες με κενο αναμεσα σε κ ά θ ε
#     γραμμα (π.χ. "Σ Υ Ν Ο Λ Ο" αντι "ΣΥΝΟΛΟ"), κατι που εκανε αδυνατη
#     καθε αναζητηση λεξης-κλειδιου. Διορθωνεται με αφαιρεση κενων
#     αναμεσα σε μεμονωμενα γραμματα.
#  3) Πολυ μεγαλυτερη λιστα ετικετων για ποσο (ΠΛΗΡΩΤΕΟ, ΚΑΘΑΡΟ ΠΟΣΟ,
#     ΚΑΘΑΡ.ΔΙΚΑΙΟΥΧΟΥ, ΣΥΝΟΛΟ ΕΝΤΑΛΜΑΤΟΣ, ΣΥΝ.ΕΝΤ/ΤΟΣ κλπ) με προσοχη
#     να ΜΗΝ μπερδευεται το "ΣΥΝΟΛΟ ΚΡΑΤΗΣΕΩΝ" (κρατησεις, οχι ποσο) και
#     η "ΚΑΘΑΡΗ ΑΞΙΑ" (αξια προ ΦΠΑ -- διαφορετικη εννοια απο καθαρο
#     ποσο μετα κρατησεων!).
#  4) Αναδοχος: 5 στρατηγικες -- (Α) αγκυρα σε ΑΦΜ-9 ψηφια στην ιδια
#     γραμμη, (Β) αναζητηση νομικης μορφης Ο.Ε./Α.Ε./Ε.Π.Ε./Ι.Κ.Ε. σε
#     ολο το εγγραφο (πιανει περιπτωσεις πολυ μακρια απο την ετικετα
#     ΔΙΚΑΙΟΥΧΟΣ), (Β2) μοτιβο "ΟΝΟΜΑ ΕΠΩΝΥΜΟ ΚΑ 30.6261.05" συχνο σε
#     παλια εντυπα, (Γ) παλιος πινακας με "|", (Δ) γειτονια ετικετας
#     ΔΙΚΑΙΟΥΧ με ευρυτερο παραθυρο. Σε καθε στρατηγια αποκλειουμε:
#     σκουπιδια/συμβολα, boilerplate φρασεις (ΕΛΛΗΝΙΚΗ ΔΗΜΟΚΡΑΤΙΑ,
#     ΝΟΜΟΣ, ΤΑΜΕΙΑΚΗ ΥΠΗΡΕΣΙΑ κλπ), περιπτωσεις που ο "αναδοχος" =
#     ο ιδιος ο φορεας, και περιπτωσεις που ταυτιζεται με το subject
#     (= επαναλαμβανομενη περιγραφη, οχι ονομα).
#
# Δοκιμασμενο σε 5 PDF διαφορετικων δημων/ετων/προτυπων: 5/5 σωστα.
# ============================================================
import requests
import io
import pypdf
import os
import time
import pandas as pd
import re

input_file = 'diavgeia_scraped_full_data.csv'
output_file = 'diavgeia_final_extracted_v3.csv'   # νεο αρχειο -- κρατα το v2/προηγουμενο για συγκριση

df_payments = pd.read_csv(input_file)
required_cols = {'ada', 'documentUrl'}
missing = required_cols - set(df_payments.columns)
if missing:
    raise ValueError(f"Λείπουν απαραίτητες στήλες απο το {input_file}: {missing}.")
print(f"Φορτώθηκαν {len(df_payments)} εγγραφές. Στήλες: {list(df_payments.columns)}")

GREEK_FIX = str.maketrans('ΆΈΉΊΪΌΎΫΏ∆\u2126', 'ΑΕΗΙΙΟΥΥΩΔΩ')
MONEY = r'(\d{1,3}(?:\.\d{3})*,\d{2})'

def normalize(text):
    txt = text.upper().translate(GREEK_FIX)
    txt = re.sub(r'[ \t]+', ' ', txt)
    # Διορθωση "διαστιχου": "Χ Ρ Η Μ Α" -> "ΧΡΗΜΑ"
    txt = re.sub(r'(?:[Α-ΩΪΫ] ){2,}[Α-ΩΪΫ]', lambda m: m.group(0).replace(' ', ''), txt)
    return txt

def find_amount_near_label(flat_txt, label_pattern, window=70, exclude_after=None):
    candidates = []
    for m in re.finditer(label_pattern, flat_txt):
        if exclude_after and re.match(exclude_after, flat_txt[m.end(): m.end() + 15]):
            continue
        after = list(re.finditer(MONEY, flat_txt[m.end(): m.end() + window]))
        if after:
            candidates.append(after[0].group(1)); continue
        before = list(re.finditer(MONEY, flat_txt[max(0, m.start() - window): m.start()]))
        if before:
            candidates.append(before[-1].group(1))
    return candidates

NET_LABELS = [r'ΠΛΗΡΩΤΕΟ\s*ΠΟΣΟ', r'ΚΑΘΑΡΟ\s*ΠΟΣΟ', r'ΚΑΘΑΡ\.\s*ΔΙΚΑΙΟΥΧΟΥ',
              r'ΚΑΘΑΡ\.\s*ΠΟΣΟ', r'ΠΛΗΡΩΤΕΟ(?!\s*ΑΠΟ)']
GROSS_LABELS = [r'ΣΥΝΟΛΙΚΟ\s*ΠΟΣΟ', r'ΣΥΝΟΛΙΚΗ\s*ΑΞΙΑ', r'ΣΥΝΟΛΟ\s*ΕΝΤΑΛΜΑΤΟΣ',
                r'ΣΥΝ\.?\s*ΕΝΤ[\./]?\s*ΤΟΣ', r'ΣΥΝΟΛΟ(?!\s*ΚΡΑΤΗΣ)']

def extract_amounts(flat_txt):
    net, gross = [], []
    for lab in NET_LABELS:
        net += find_amount_near_label(flat_txt, lab)
    for lab in GROSS_LABELS:
        excl = r'\s*ΚΡΑΤΗΣ' if lab.startswith('ΣΥΝΟΛΟ(') else None
        gross += find_amount_near_label(flat_txt, lab, exclude_after=excl)
    return (net[-1] if net else None), (gross[-1] if gross else None)

def extract_assignment_type(flat_txt):
    if "ΑΠΕΥΘΕΙΑΣ ΑΝΑΘΕΣ" in flat_txt: return "Απευθείας Ανάθεση"
    if "ΣΥΝΟΠΤΙΚ" in flat_txt: return "Συνοπτικός Διαγωνισμός"
    if "ΑΝΟΙΚΤ" in flat_txt or "ΑΝΟΙΧΤ" in flat_txt: return "Ανοικτός Διαγωνισμός"
    if "ΠΡΟΧΕΙΡ" in flat_txt: return "Πρόχειρος Διαγωνισμός"
    if "ΔΙΑΓΩΝΙΣΜ" in flat_txt: return "Διαγωνισμός"
    if "ΑΝΑΘΕΣΗΣ" in flat_txt or "ΑΝΑΘΕΣΗ" in flat_txt: return "Ανάθεση (μη προσδιορισμένος τύπος)"
    return "Απλή Πληρωμή / Τακτική"

LABEL_PREFIXES = [r'^ΕΠΩΝΥΜΙΑ\.*\s*:?\s*', r'^ΟΝΟΜ/ΜΟ\s*Η\s*ΕΠΩΝΥΜΙΑ\.*\s*:?\s*',
                   r'^ΟΝΟΜΑΤΕΠΩΝΥΜΟ\.*\s*:?\s*', r'^ΔΙΚΑΙΟΥΧ\w*\.*\s*:?\s*']
def strip_label_prefix(s):
    for pat in LABEL_PREFIXES:
        s = re.sub(pat, '', s)
    return s.strip(' .:')

def _alnum(s):
    return re.sub(r'[^Α-ΩΪΫA-Z0-9]', '', s)

def overlaps_subject(cand, subject_norm):
    if not subject_norm or len(cand) < 6:
        return False
    c, s = _alnum(cand), _alnum(subject_norm)
    if not c or not s:
        return False
    return c in s or s in c or (len(c) > 15 and c[:15] in s)

BOILERPLATE = ["ΔΙΚΑΙΟΥΧ", "ΠΡΟΙΣΤΑΜΕΝ", "ΠΡ/ΝΟΣ", "ΛΟΓΙΣΤΗΡ", "ΔΗΜΑΡΧΟΣ", "ΤΑΜΙΑΣ", "ΑΘΕΩΡΗΤΟ",
               "ΑΙΤΙΑ", "ΠΛΗΡΩΜ", "ΠΟΣΟ", "ΕΥΡΩ", "ΕΝΤΑΛΜΑ", "ΟΛΟΓΡΑΦ", "ΑΡΙΘΜΗΤ", "ΚΡΑΤΗΣ", "ΣΥΝΟΛ",
               "ΤΑΜΕΙΑΚΗ ΥΠΗΡΕΣΙΑ", "ΠΛΗΡΩΤΕΟ ΑΠΟ", "ΑΙΤΙΟΛΟΓ", "ΧΡΗΜΑΤΟΔΟΤΗΣ", "ΚΑΤΑΣΤΑΣΗ ΠΛΗΡΩ",
               "ΕΛΛΗΝΙΚΗ ΔΗΜΟΚΡΑΤΙΑ", "ΝΟΜΟΣ ", "ΠΕΡΙΦΕΡΕΙΑ ", "ΟΙΚΟΝ", "ΧΡΗΣΗ ΤΟΥ ΕΤΟΥΣ"]

def is_garbage(cand, subject_norm=''):
    if len(cand) < 4: return True
    letters = sum(1 for c in cand if c.isalpha())
    if letters / max(len(cand), 1) < 0.5: return True
    if sum(c.isdigit() for c in cand) > 2: return True
    if any(w in cand for w in BOILERPLATE): return True
    if overlaps_subject(cand, subject_norm): return True
    return False

LEGAL_SUFFIXES = [r'\bΟ\.?\s?Ε\.?(?:\s|$)', r'\bΑ\.?\s?Ε\.?(?:\s|$)', r'\bΕ\.?\s?Π\.?\s?Ε\.?(?:\s|$)',
                   r'\bΙ\.?\s?Κ\.?\s?Ε\.?(?:\s|$)', r'ΚΟΙΝΟΠΡΑΞΙΑ', r'ΜΟΝΟΠΡΟΣΩΠΗ']
NAME_PLUS_BUDGETCODE = re.compile(r'^([Α-Ω]+(?:\s[Α-Ω]+){0,3})\s*[\d.,]*\s*ΚΑ\s+\d')

def extract_contractor(flat_txt, lines, foreas_name, subject=''):
    foreas_up = (foreas_name or '').strip().upper()
    subj_norm = re.sub(r'\s+', ' ', normalize(subject or ''))

    m = re.search(r'Α\.?Φ\.?Μ\.?\s*:?\s*([Α-ΩΪΫA-Z0-9\.\-\s&]{4,80}?)\s*(\d{9})\b', flat_txt)
    if m:
        name = strip_label_prefix(m.group(1).strip(' .-'))
        if not is_garbage(name, subj_norm) and name.upper() != foreas_up:
            return name

    for line in lines:
        if any(re.search(suf, line) for suf in LEGAL_SUFFIXES):
            cand = strip_label_prefix(line)
            cand = re.split(r'\s{2,}|,\s*ΚΑ\s|\bΚΑ\s\d', cand)[0].strip()
            if not is_garbage(cand, subj_norm) and cand.upper() != foreas_up and len(cand) <= 80:
                return cand

    for line in lines:
        m2 = NAME_PLUS_BUDGETCODE.match(line)
        if m2:
            cand = m2.group(1).strip()
            if not is_garbage(cand, subj_norm) and cand.upper() != foreas_up:
                return cand

    for i, line in enumerate(lines):
        if 'ΔΙΚΑΙΟΥΧ' in line and '|' in line:
            for j in range(i + 1, min(i + 6, len(lines))):
                row = lines[j]
                if not row.strip('-| ') or '|' not in row:
                    continue
                cell = row.split('|')[1].strip() if row.startswith('|') else row.split('|')[0].strip()
                if cell and not is_garbage(cell, subj_norm) and cell.upper() != foreas_up:
                    return cell
            break

    for i, line in enumerate(lines):
        if 'ΔΙΚΑΙΟΥΧ' in line:
            for idx in range(max(0, i - 10), min(len(lines), i + 10)):
                cand = strip_label_prefix(lines[idx].strip())
                if is_garbage(cand, subj_norm) or cand.upper() == foreas_up:
                    continue
                return cand
    return "Δεν εντοπίστηκε"

def process_pdf_text(raw_text, foreas_name='', subject=''):
    txt = normalize(raw_text)
    flat = re.sub(r'\s+', ' ', txt)
    lines = [l.strip() for l in txt.split('\n') if l.strip()]
    net, gross = extract_amounts(flat)
    return {
        'amount_net': net,
        'amount_gross': gross,
        'assignment_type': extract_assignment_type(flat),
        'contractor_name': extract_contractor(flat, lines, foreas_name, subject)
    }

if os.path.exists(output_file):
    df_progress = pd.read_csv(output_file)
    processed_adas = set(df_progress['ada'].tolist())
    results = df_progress.to_dict('records')
    print(f"📦 Συνέχεια από checkpoint: {len(processed_adas)} έγγραφα ήδη έτοιμα.")
else:
    processed_adas = set()
    results = []
    print("🚀 Πλήρης επανεπεξεργασία όλων των PDF με την τελική, δοκιμασμένη λογική...")

headers = {'User-Agent': 'Mozilla/5.0'}
total_docs = len(df_payments)

for index, row in df_payments.iterrows():
    ada = row['ada']
    if ada in processed_adas:
        continue
    url = row['documentUrl']
    foreas_name = row.get('foreas_name', '') if hasattr(row, 'get') else ''
    subject = row.get('subject', '') if hasattr(row, 'get') else ''
    fields = {'amount_net': "Δεν βρέθηκε", 'amount_gross': "Δεν βρέθηκε",
              'assignment_type': "Σφάλμα", 'contractor_name': "Σφάλμα"}
    try:
        response = requests.get(url, headers=headers, timeout=10)
        if response.status_code == 200:
            with io.BytesIO(response.content) as f:
                reader = pypdf.PdfReader(f)
                text = "".join([p.extract_text() or "" for p in reader.pages])
                fields = process_pdf_text(text, foreas_name, subject)
        else:
            fields = {'amount_net': "Σφάλμα Λήψης", 'amount_gross': "Σφάλμα Λήψης",
                       'assignment_type': "Σφάλμα", 'contractor_name': "Σφάλμα"}
    except Exception:
        fields = {'amount_net': "Σφάλμα Ανάγνωσης", 'amount_gross': "Σφάλμα Ανάγνωσης",
                   'assignment_type': "Σφάλμα", 'contractor_name': "Σφάλμα"}

    row_dict = row.to_dict()
    row_dict.update(fields)
    results.append(row_dict)
    processed_adas.add(ada)

    if len(results) % 50 == 0:
        pd.DataFrame(results).to_csv(output_file, index=False, encoding='utf-8-sig')
        print(f"⏳ {len(results)}/{total_docs} | net: {fields['amount_net']} | ανάδοχος: {str(fields['contractor_name'])[:35]}")
        time.sleep(0.3)

pd.DataFrame(results).to_csv(output_file, index=False, encoding='utf-8-sig')
print("✅ Ολοκληρώθηκε το τελικό rebuild!")
```

```python
# ============================================================
# ΚΑΘΑΡΙΣΜΟΣ & ΤΕΛΙΚΟ DATASET
#
# 1) Καθαρισμος "ψευτο-θετικων" στον αναδοχο: μερικες φορες ο κωδικας
#    εβρισκε κατι (οχι "Δεν εντοπίστηκε") αλλα ηταν σκουπιδι (π.χ.
#    "ΒΕΒΑΙΩΣΗ ΚΙΝΗΣΗΣ", "ΑΣΦΑΛΙΣΤΙΚΗ ΕΝΗΜΕΡΟΤΗΤΑ", κομματια λεξεων).
#    Αυτα τα ξανακανουμε "Δεν εντοπίστηκε" (ειλικρινης αγνοια > ψευτος
#    θετικος). Δεν χρειαζεται να ξαναβγουμε στα PDF -- δουλευουμε πανω
#    στο ηδη υπαρχον CSV.
# 2) "Αξιοπιστο ποσο" = εχει ΚΑΙ καθαρο ΚΑΙ συνολικο ποσο, και τα δυο
#    περνανε τον λογικο ελεγχο (καθαρο <= συνολικο)
# ============================================================
import pandas as pd
import re

df = pd.read_csv('diavgeia_final_extracted_v3.csv')
print(f"Ξεκιναμε με {len(df)} εγγραφες.")

# --- Καθαρισμος "ψευτο-θετικων" στο ονομα αναδοχου ---
GARBAGE_PATTERNS = ['ΤΑΚΤΙ', 'Κ.Α.Ε', 'ΑΣΦΑΛΙΣΤΙΚΗ ΕΝΗΜΕΡΟΤΗΤΑ', 'ΒΕΒΑΙΩΣΗ', 'ΚΑΘ ΑΞΙΑ',
                    'ΚΑΘ. ΑΞΙΑ', 'ΚΑΘΑΡΗ ΑΞΙΑ', 'ΠΕΡΙΓΡΑΦΗ ΛΟΓΑΡΙΑΣΜΟΥ', 'ΚΑΤΗΓ',
                    'ΔΙΚΑΙΟΛΟΓΗΤΙΚΑ', 'ΛΟΓΑΡΙΑΣΜΟΥ', 'ΠΕΡΙΓΡΑΦΗ', 'Φ.Π.Α',
                    'ΥΠΕΡ ΤΡΙΤΩΝ', 'ΕΛΛΗΝΙΚΗ ΔΗΜΟΚΡ', 'ΕΛΛΗΝΙΚΗ ΔΗΜΟΚΡΑΤΙΑ',
                    'ΠΡΟΥΠΟΛΟΓΙΣΜΟ', 'ΦΥΛΛΟ ΕΚΚΑΘΑΡΙΣΗΣ', 'ΣΕΙΡΑ : Α', 'ΣΕΙΡΑ :Α',
                    # Επιπλεον garbage που εντοπιστηκε μετα τον ελεγχο του dataset:
                    'ΕΚΘ. ΑΝΑΛ', 'ΕΚΘΕΣΗ ΑΝΑΛ', 'ΑΝΑΛΥΣΗΣ ΔΑΠΑΝΗΣ',  # επικεφαλιδες πινακα
                    'ΚΩΔΙΚΟΣ ΔΙΑΜΟΡΦΩΘΕΝΤΑ', 'ΕΝΤΑΛΘΕΝΤΑ ΜΕ ΤΟ ΠΑΡΟΝ',  # επικεφαλιδα πινακα
                    'ΕΛΕΓΚΤΙΚΟ ΣΥΝΕΔΡΙΟ',   # boilerplate κειμενου
                    'ΔΗΜΟΣ ΚΕΝΤΡΙΚΗΣ ΚΕΡΚΥΡ', 'ΔΗΜΟΣ ΑΡΓΟΥΣ', 'ΔΗΜΟΣ ΝΙΚΑΙΑΣ',
                    'ΔΗΜΟΣ ΓΛΥΦΑΔ', 'ΔΗΜΟΣ ΧΑΝΙ', 'ΔΗΜΟΣ ΠΑΥΛΟΥ',  # δημοι ως αναδοχοι = λαθος
                    '|SA |', 'ΔΗΜ. ΣΥΜΒ.',  # κομματια πινακα
                    'ΚΩΔ.-ΛΟΓ/ΣΜΟΣ', 'ΛΟΓ/ΣΜΟΣ-ΠΑΡ/ΚΟ',  # επικεφαλιδες πινακα
                    'ΑΠΟΦΑΣΗ ΑΝΑΘΕΣΗΣ ΔΗΜΑΡΧΟΥ', 'ΑΠΟΦΑΣΗ ΔΗΜΑΡΧΟΥ',  # boilerplate
                    'ΤΑΜΕΙΟ ΠΑΡ', 'ΠΑΡΑΚΑΤΑΘΗΚΩΝ']  # Ταμειο Παρακαταθηκων, οχι εργολαβος

def clean_contractor(name):
    if pd.isna(name) or str(name).strip() == '':
        return 'Δεν εντοπίστηκε'
    name_up = str(name).upper()
    if any(p in name_up for p in GARBAGE_PATTERNS):
        return 'Δεν εντοπίστηκε'
    words = str(name).split()
    if len(words) == 1 and len(str(name)) < 10 and not re.search(r'(Ο\.?Ε\.?|Α\.?Ε\.?|ΕΠΕ|ΙΚΕ)$', name_up):
        return 'Δεν εντοπίστηκε'
    return name

before_found = (df['contractor_name'] != 'Δεν εντοπίστηκε').sum()
df['contractor_name'] = df['contractor_name'].apply(clean_contractor)
after_found = (df['contractor_name'] != 'Δεν εντοπίστηκε').sum()
print(f"Ανάδοχος -- πριν καθαρισμα: {before_found}, μετα: {after_found} "
      f"(αφαιρεθηκαν {before_found - after_found} 'ψευτο-θετικα')")

# --- Μετατροπη ποσων σε πραγματικους αριθμους ---
def to_float(s):
    try:
        return float(str(s).replace('.', '').replace(',', '.'))
    except (ValueError, TypeError):
        return None

df['amount_net_eur'] = df['amount_net'].apply(to_float)
df['amount_gross_eur'] = df['amount_gross'].apply(to_float)

# --- Ετος / Μηνας απο το issueDate ---
df['issue_date_parsed'] = pd.to_datetime(df['issueDate'], format='%d/%m/%Y %H:%M:%S', errors='coerce')
df['year'] = df['issue_date_parsed'].dt.year
df['month'] = df['issue_date_parsed'].dt.month

# --- Σημαια αξιοπιστιας ποσου ---
# Χρησιμοποιουμε ΜΟΝΟ το καθαρο/πληρωτεο ποσο (amount_net_eur) ως κυριο ποσο.
# Λογοι: (1) εξαγεται πιο αξιοπιστα (η ετικετα "ΠΛΗΡΩΤΕΟ ΠΟΣΟ" εμφανιζεται
# συνεπως σε ολα τα templates), (2) ειναι αυτο που πραγματικα πληρωθηκε,
# (3) το gross σε ~55% των εγγραφων ταυτιζεται με το net λογω συγχυσης
# στην εξαγωγη (διμερης στηλη στο PDF αναποδογυριζει τη σειρα των τιμων).
ok_download = df['assignment_type'] != 'Σφάλμα'
has_net = df['amount_net_eur'].notna()
df['reliable_amount'] = ok_download & has_net

print(f"\nΑξιοπιστο ποσο (εχει καθαρο/πληρωτεο ποσο): {df['reliable_amount'].sum()} / {len(df)}")
print(f"Εχουν επιπλεον και (καθαρισμενο) ανάδοχο: {(df['reliable_amount'] & (df['contractor_name']!='Δεν εντοπίστηκε')).sum()}")

# --- Κραταμε μονο τις αξιοπιστες ---
df_clean = df[df['reliable_amount']].copy()
before = len(df_clean)

# Βημα Α: Αφαιρεση μηδενικων ποσων (σφαλματα εξαγωγης)
df_clean = df_clean[df_clean['amount_net_eur'] > 0]
print(f"Αφαιρεθηκαν μηδενικα ποσα: {before - len(df_clean)}")

# Βημα Β: Αφαιρεση κεντρικων φορεων (οχι δημοι -- εκτος θεματος)
before2 = len(df_clean)
df_clean = df_clean[~df_clean['foreas_name'].str.contains('ΚΤΙΡΙΑΚΕΣ ΥΠΟΔΟΜΕΣ', na=False)]
print(f"Αφαιρεθηκαν κεντρικοι φορεις (ΚΤΙΡΙΑΚΕΣ ΥΠΟΔΟΜΕΣ ΑΕ): {before2 - len(df_clean)}")

# Βημα Γ: Αφαιρεση ποσων < 1€ (τραπεζικα εξοδα/σφαλματα, οχι πραγματικες εργασιες)
before3 = len(df_clean)
df_clean = df_clean[df_clean['amount_net_eur'] >= 1.0]
print(f"Αφαιρεθηκαν ποσα < 1€ (σφαλματα): {before3 - len(df_clean)}")

# Βημα Δ: Αφαιρεση διπλοτυπων ΑΔΑ
before4 = len(df_clean)
df_clean = df_clean.drop_duplicates(subset=['ada'])
print(f"Αφαιρεθηκαν διπλοτυπα ΑΔΑ: {before4 - len(df_clean)}")

# Βημα Ε: Αφαιρεση διπλοτυπων (ιδιος δημος + ποσο + ημερομηνια = ανατυπωμενα εντυπα)
before5 = len(df_clean)
df_clean = df_clean.drop_duplicates(subset=['foreas_name', 'amount_net_eur', 'issueDate'], keep='first')
print(f"Αφαιρεθηκαν διπλοτυπα (δημος+ποσο+ημερομηνια): {before5 - len(df_clean)}")

print(f"\nΣυνολο αφαιρεθεντων: {before - len(df_clean)} | Απομενουν: {len(df_clean)}")

# --- Κραταμε μονο τις χρησιμες στηλες για το τελικο dataset ---
final_cols = ['ada', 'issueDate', 'year', 'month', 'subject', 'foreas_name',
              'organization', 'decisionType', 'amount_net_eur',
              'assignment_type', 'contractor_name', 'documentUrl']
df_final = df_clean[final_cols].rename(columns={
    'subject': 'titlos_apofasis',
    'amount_net_eur': 'poso_eur',   # καθαρο/πληρωτεο ποσο -- η μοναδικη στηλη ποσου
})

df_final.to_csv('diavgeia_TELIKO_dataset.csv', index=False, encoding='utf-8-sig')
print(f"\n✅ Τελικο dataset: {len(df_final)} εγγραφες -> diavgeia_TELIKO_dataset.csv")
print(f"   (ελαχιστο απαιτουμενο απο την εκφωνηση: 500)")
print(f"\nΚατανομη ανα ετος:")
print(df_final['year'].value_counts().sort_index())
print(f"\nΤυχαιο δειγμα 8 εγγραφων για χειροκινητο ελεγχο:")
display(df_final.sample(8, random_state=42)[['ada','titlos_apofasis','poso_eur','contractor_name','year','documentUrl']])
```

```python
# ============================================================
# ΑΝΑΛΥΣΗ ΔΕΔΟΜΕΝΩΝ ΣΕ PYTHON
#
# - groupby ανα ετος, μηνα, δημο, αναδοχο
# - νεες στηλες που βοηθουν την αναλυση (κατηγορια, size_category)
# - μεγαλυτερες αποφασεις
# - συχνοτερες λεξεις στους τιτλους
# ============================================================
import pandas as pd
import re
from collections import Counter

df = pd.read_csv('diavgeia_TELIKO_dataset.csv')
print(f"Dataset: {len(df)} εγγραφες | Περιοδος: {int(df['year'].min())}-{int(df['year'].max())}")
print(f"Συνολο αξια: €{df['poso_eur'].sum():,.0f} | Μεσος ορος: €{df['poso_eur'].mean():,.0f} | Διαμεσος: €{df['poso_eur'].median():,.0f}")

# --- Νεα στηλη: κατηγορια αποφασης ---
def categorize(title):
    t = str(title).upper()
    if any(w in t for w in ['ΑΝΑΚΑΙΝΙΣ', 'ΑΝΑΒΑΘΜΙΣ', 'ΕΝΕΡΓΕΙΑΚ']):
        return 'Ανακαίνιση / Ενεργειακή Αναβάθμιση'
    if any(w in t for w in ['ΕΠΙΣΚΕΥ', 'ΑΠΟΚΑΤΑΣ']):
        return 'Επισκευή'
    if any(w in t for w in ['ΣΥΝΤΗΡ']):
        return 'Συντήρηση'
    if any(w in t for w in ['ΠΡΟΜΗΘ']):
        return 'Προμήθεια Υλικών'
    if any(w in t for w in ['ΕΠΙΧΟΡΗΓ', 'ΧΡΗΜΑΤΟΔΟΤ', 'ΑΠΟΔΟΣΗ']):
        return 'Επιχορήγηση / Χρηματοδότηση'
    if any(w in t for w in ['ΜΙΣΘ', 'ΑΜΟΙΒ']):
        return 'Αμοιβή / Μισθοδοσία'
    return 'Λοιπά'

df['kategoria'] = df['titlos_apofasis'].apply(categorize)

# --- Νεα στηλη: μεγεθος πληρωμης ---
def size_cat(v):
    if v < 1000: return 'Α. Κάτω από €1.000'
    if v < 5000: return 'Β. €1.000 - €5.000'
    if v < 20000: return 'Γ. €5.000 - €20.000'
    if v < 50000: return 'Δ. €20.000 - €50.000'
    return 'Ε. Άνω από €50.000'

df['size_category'] = df['poso_eur'].apply(size_cat)

# Αποθηκευση του εμπλουτισμενου dataset
df.to_csv('diavgeia_TELIKO_dataset.csv', index=False, encoding='utf-8-sig')
print("\n✅ Dataset εμπλουτιστηκε με στηλες 'kategoria' και 'size_category'")

print("\n" + "="*55)
print("1. ΤΑΣΕΙΣ ΑΝΑ ΕΤΟΣ (2014-2025)")
print("="*55)
by_year = df[df['year'] <= 2025].groupby('year').agg(
    apofaseis=('ada', 'count'),
    synoliko_poso_eur=('poso_eur', 'sum'),
    meso_poso_eur=('poso_eur', 'mean')
).round(0).astype(int)
display(by_year)

print("\n--- Βασικα ευρηματα τασεων (για αρθρο) ---")
poso_2014 = by_year.loc[2014, 'synoliko_poso_eur']
poso_2023 = by_year.loc[2023, 'synoliko_poso_eur']
meso_2015 = by_year.loc[2015, 'meso_poso_eur']
meso_2024 = by_year.loc[2024, 'meso_poso_eur']
print(f"Αυξηση 2014→2023: {(poso_2023 - poso_2014) / poso_2014 * 100:.0f}%")
print(f"Μεσο ποσο 2015: €{meso_2015:,}")
print(f"Μεσο ποσο 2024: €{meso_2024:,}")
print(f"Πολλαπλασιασμος μεσου ποσου: {meso_2024/meso_2015:.1f}x")
print(f"Μεσος ορος δαπανης ανα ετος (2014-2026): €{df['poso_eur'].sum()/13:,.0f}")

# Εξαγωγη CSV για γραφημα 1 (ποσο ανα ετος)
by_year_csv = df[df['year'].between(2014,2026)].groupby('year').agg(
    Apofaseis=('ada','count'),
    Synoliko_poso_EUR=('poso_eur','sum')
).reset_index()
by_year_csv.columns = ['Etos','Apofaseis','Synoliko_poso_EUR']
by_year_csv['Synoliko_poso_EUR'] = by_year_csv['Synoliko_poso_EUR'].round(0).astype(int)
by_year_csv['Katastasi'] = by_year_csv['Etos'].apply(
    lambda y: 'Ατελές έτος (έως Μάιο 2026)' if y == 2026 else 'Πλήρες έτος'
)
by_year_csv.to_csv('grafima1_poso_ana_etos.csv', index=False, encoding='utf-8-sig')
print("CSV αποθηκευτηκε: grafima1_poso_ana_etos.csv")

print("\n" + "="*55)
print("2. TOP 15 ΔΗΜΟΙ ΑΝΑ ΣΥΝΟΛΙΚΟ ΠΟΣΟ")
print("="*55)
top_foreas_poso = df.groupby('foreas_name').agg(
    apofaseis=('ada', 'count'),
    synoliko_poso_eur=('poso_eur', 'sum')
).sort_values('synoliko_poso_eur', ascending=False).head(15)
top_foreas_poso['synoliko_poso_eur'] = top_foreas_poso['synoliko_poso_eur'].round(0).astype(int)
display(top_foreas_poso)

# Εξαγωγη CSV για γραφημα 2 (top 15 δημοι)
top15_csv = df.groupby('foreas_name').agg(
    Apofaseis=('ada','count'),
    Synoliko_poso_EUR=('poso_eur','sum')
).sort_values('Synoliko_poso_EUR', ascending=False).head(15).reset_index()
top15_csv.columns = ['Dimos','Apofaseis','Synoliko_poso_EUR']
top15_csv['Dimos'] = top15_csv['Dimos'].str.replace('ΔΗΜΟΣ ','',regex=False).str.title()
top15_csv['Synoliko_poso_EUR'] = top15_csv['Synoliko_poso_EUR'].round(0).astype(int)
top15_csv[::-1].reset_index(drop=True).to_csv(
    'grafima2_top15_dimoi.csv', index=False, encoding='utf-8-sig')
print("CSV αποθηκευτηκε: grafima2_top15_dimoi.csv")

print("\n" + "="*55)
print("3. TOP 15 ΔΗΜΟΙ ΑΝΑ ΑΡΙΘΜΟ ΑΠΟΦΑΣΕΩΝ")
print("="*55)
top_foreas_count = df.groupby('foreas_name').agg(
    apofaseis=('ada', 'count'),
    synoliko_poso_eur=('poso_eur', 'sum')
).sort_values('apofaseis', ascending=False).head(15)
top_foreas_count['synoliko_poso_eur'] = top_foreas_count['synoliko_poso_eur'].round(0).astype(int)
display(top_foreas_count)

print("\n" + "="*55)
print("4. ΜΗΝΙΑΙΑ ΚΑΤΑΝΟΜΗ ΑΠΟΦΑΣΕΩΝ (ολα τα ετη)")
print("="*55)
months_gr = {1:'Ιαν',2:'Φεβ',3:'Μαρ',4:'Απρ',5:'Μαι',6:'Ιουν',
             7:'Ιουλ',8:'Αυγ',9:'Σεπ',10:'Οκτ',11:'Νοε',12:'Δεκ'}
by_month = df.groupby('month').agg(
    apofaseis=('ada', 'count'),
    synoliko_poso_eur=('poso_eur', 'sum')
).round(0)
by_month.index = by_month.index.map(months_gr)
display(by_month)

# Εξαγωγη CSV για γραφημα 3 (μηνιαια κατανομη -- ξεφορτωμα προυπολογισμου)
by_month_csv = df.groupby('month').agg(
    Apofaseis=('ada', 'count'),
    Synoliko_poso_EUR=('poso_eur', 'sum')
).reset_index()
by_month_csv['Month'] = by_month_csv['month'].map(months_gr)
by_month_csv['Synoliko_poso_EUR'] = by_month_csv['Synoliko_poso_EUR'].round(0).astype(int)
by_month_csv['Katastasi'] = by_month_csv['month'].apply(
    lambda m: 'Δεκέμβριος (peak)' if m == 12 else 'Υπόλοιποι μήνες'
)
by_month_csv[['Month','Apofaseis','Synoliko_poso_EUR','Katastasi']].to_csv(
    'grafima3_mhniaia_katanom.csv', index=False, encoding='utf-8-sig')
print("\n--- Βασικα ευρηματα μηνιαιας κατανομης (για αρθρο) ---")
dec = by_month_csv[by_month_csv['month']==12]['Synoliko_poso_EUR'].values[0]
jan = by_month_csv[by_month_csv['month']==1]['Synoliko_poso_EUR'].values[0]
jul = by_month_csv[by_month_csv['month']==7]['Synoliko_poso_EUR'].values[0]
print(f"Δεκεμβριος: €{dec:,} | Ιουλιος: €{jul:,} | Ιανουαριος: €{jan:,}")
print(f"Δεκεμβριος vs Ιανουαριος: {dec/jan:.0f}x μεγαλυτερος")
print(f"CSV αποθηκευτηκε: grafima3_mhniaia_katanom.csv")

print("\n" + "="*55)
print("5. ΚΑΤΑΝΟΜΗ ΑΝΑ ΚΑΤΗΓΟΡΙΑ ΑΠΟΦΑΣΗΣ")
print("="*55)
by_kat = df.groupby('kategoria').agg(
    apofaseis=('ada', 'count'),
    synoliko_poso_eur=('poso_eur', 'sum')
).sort_values('synoliko_poso_eur', ascending=False)
by_kat['synoliko_poso_eur'] = by_kat['synoliko_poso_eur'].round(0).astype(int)
display(by_kat)

print("\n" + "="*55)
print("6. ΚΑΤΑΝΟΜΗ ΑΝΑ ΜΕΓΕΘΟΣ ΠΛΗΡΩΜΗΣ")
print("="*55)
by_size = df.groupby('size_category').agg(
    apofaseis=('ada', 'count'),
    synoliko_poso_eur=('poso_eur', 'sum')
).sort_index()
by_size['synoliko_poso_eur'] = by_size['synoliko_poso_eur'].round(0).astype(int)
display(by_size)

print("\n--- Βασικα ευρηματα μεγεθους (για αρθρο) ---")
total_apofaseis = by_size['apofaseis'].sum()
total_poso = by_size['synoliko_poso_eur'].sum()

mikra = by_size.loc['Α. Κάτω από €1.000']
megala = by_size.loc['Ε. Άνω από €50.000']

print(f"Εντολες <€1.000: {mikra['apofaseis']} ({mikra['apofaseis']/total_apofaseis:.1%} του πληθους)")
print(f"Ποσο <€1.000:    €{mikra['synoliko_poso_eur']:,} ({mikra['synoliko_poso_eur']/total_poso:.1%} του συνολου)")
print(f"Εντολες >€50.000: {megala['apofaseis']} ({megala['apofaseis']/total_apofaseis:.1%} του πληθους)")
print(f"Ποσο >€50.000:    €{megala['synoliko_poso_eur']:,} ({megala['synoliko_poso_eur']/total_poso:.1%} του συνολου)")

print("\n" + "="*55)
print("7. ΟΙ 10 ΜΕΓΑΛΥΤΕΡΕΣ ΑΠΟΦΑΣΕΙΣ")
print("="*55)
top10 = df.nlargest(10, 'poso_eur')[['ada','year','titlos_apofasis','foreas_name','poso_eur','contractor_name']]
top10['poso_eur'] = top10['poso_eur'].round(0).astype(int)
display(top10)

print("\n" + "="*55)
print("8. TOP 15 ΑΝΑΔΟΧΟΙ (σε εγγραφες με αναδοχο)")
print("="*55)
has_contractor = df[df['contractor_name'] != 'Δεν εντοπίστηκε']
top_contractors = has_contractor.groupby('contractor_name').agg(
    anatheseis=('ada', 'count'),
    synoliko_poso_eur=('poso_eur', 'sum')
).sort_values('synoliko_poso_eur', ascending=False).head(15)
top_contractors['synoliko_poso_eur'] = top_contractors['synoliko_poso_eur'].round(0).astype(int)
display(top_contractors)

print("\n" + "="*55)
print("9. ΣΥΧΝΟΤΕΡΕΣ ΛΕΞΕΙΣ ΣΤΟΥΣ ΤΙΤΛΟΥΣ (top 20)")
print("="*55)
stopwords = {'ΚΑΙ','ΤΟΥ','ΤΗΣ','ΤΩΝ','ΤΟ','ΤΑ','ΤΗΝ','ΤΟΝ','ΣΤΟ','ΣΤΑ','ΣΤΗ','ΣΤΟΝ',
             'ΑΠΟ','ΓΙΑ','ΜΕ','ΩΣ','ΤΕ','ΕΠΙ','ΠΡΟ','ΕΙΣ','ΑΝΑ','ΔΕ','ΜΗ','ΤΕ',
             'ΕΝ','Η','Ο','ΟΙ','ΣΕ','ΤΕ','ΝΑ','ΤΙ','Α','Β','Γ'}
all_words = []
for title in df['titlos_apofasis'].dropna():
    words = re.findall(r'[Α-ΩΪΫΆΈΉΊΌΎΏΐΰα-ωϊϋάέήίόύώ]{4,}', str(title).upper())
    all_words.extend([w for w in words if w not in stopwords])
word_freq = Counter(all_words).most_common(20)
word_df = pd.DataFrame(word_freq, columns=['λεξη', 'συχνοτητα'])
display(word_df)


print("\n" + "="*55)
print("10. ΣΥΓΚΡΙΣΗ ΔΑΠΑΝΩΝ ΠΡΙ/ΜΕΤΑ ΤΟ 2020 ΑΝΑ ΔΗΜΟ")
print("="*55)
print("Εντοπισμος δημων με εκρηκτικη αυξηση μετα το 2020")
print("ΣΗΜΕΙΩΣΗ: χρονικη συσχετιση με Ταμειο Ανακαμψης/ΕΣΠΑ 2021-2027,")
print("οχι αποδεδειγμενη αιτιακη σχεση (η πηγη δεν αναγραφεται στα εγγραφα)")

pre_2020 = df[df['year'] < 2020].groupby('foreas_name')['poso_eur'].sum()
post_2020 = df[df['year'] >= 2020].groupby('foreas_name')['poso_eur'].sum()
compare = pd.DataFrame({'pre_2020': pre_2020, 'post_2020': post_2020}).dropna()
compare = compare[compare['pre_2020'] > 10000]
compare['auxisi_pct'] = ((compare['post_2020'] - compare['pre_2020']) / compare['pre_2020'] * 100).round(0)
top_growth = compare.sort_values('auxisi_pct', ascending=False).head(10).copy()
top_growth['pre_2020'] = top_growth['pre_2020'].round(0).astype(int)
top_growth['post_2020'] = top_growth['post_2020'].round(0).astype(int)
top_growth['auxisi_pct'] = top_growth['auxisi_pct'].astype(int)
display(top_growth)

print("\n--- Εγγραφες 2026 (ατελες ετος, εως Μαιο 2026) ---")
df_2026 = df[df['year'] == 2026]
print(f"Πληθος: {len(df_2026)} εντάλματα | Συνολο: €{df_2026['poso_eur'].sum():,.0f}")

print("\n" + "="*55)
print("10. ΑΝΑΔΟΧΟΙ - ΚΑΘΑΡΗ ΑΝΑΛΥΣΗ")
print("="*55)

garbage_check = ['ΕΚΘ. ΑΝΑΛ', 'ΚΩΔΙΚΟΣ ΔΙΑΜΟΡΦ', 'ΕΝΤΑΛΘΕΝΤΑ',
                 'ΕΛΕΓΚΤΙΚΟ ΣΥΝΕΔΡ', 'ΤΑΜΕΙΟ ΠΑΡΑΚΑΤ', 'ΔΗΜΟΣ ΚΕΝΤΡΙΚΗΣ ΚΕΡΚΥΡ',
                 'ΔΗΜΟΣ ΑΡΓΟΥΣ', 'ΑΠΟΦΑΣΗ ΔΣ', 'ΣΥΝΕΡΓ ΑΣΙΑ',
                 'ΑΠΟΦΑΣΗ ΑΝΑΘΕΣΗΣ', 'ΣΧΟΛΙΚΗ ΕΠΙΤΡΟΠΗ', 'ΤΑΜΕΙΟ ΠΑΡΑ']

def is_reliable_contractor(name):
    if name == 'Δεν εντοπίστηκε':
        return False
    name_up = str(name).upper()
    return not any(g in name_up for g in garbage_check)

df['reliable_contractor'] = df['contractor_name'].apply(is_reliable_contractor)
reliable = df[df['reliable_contractor']]

print(f"Εγγραφες με αξιοπιστο αναδοχο: {len(reliable)} ({len(reliable)/len(df):.1%} του συνολου)")
print(f"Εγγραφες χωρις αναδοχο: {len(df)-len(reliable)} ({(len(df)-len(reliable))/len(df):.1%})")

print("\n--- Top 15 αναδοχοι ανα συνολικο ποσο (αξιοπιστοι) ---")
top_rel = reliable.groupby('contractor_name').agg(
    anatheseis=('ada','count'),
    synoliko_eur=('poso_eur','sum'),
    dimoi=('foreas_name', lambda x: x.nunique())
).sort_values('synoliko_eur', ascending=False).head(15)
top_rel['synoliko_eur'] = top_rel['synoliko_eur'].round(0).astype(int)
display(top_rel)

print("\n--- Κυριοι αναδοχοι ανα δημο (top 5 δημοι βασει συνολικου ποσου) ---")
top5_dimoi = df.groupby('foreas_name')['poso_eur'].sum().nlargest(5).index
for dimos in top5_dimoi:
    df_dimos = reliable[reliable['foreas_name'] == dimos]
    if len(df_dimos) == 0:
        continue
    top_an = df_dimos.groupby('contractor_name').agg(
        anatheseis=('ada','count'),
        synoliko_eur=('poso_eur','sum')
    ).sort_values('synoliko_eur', ascending=False).head(3)
    top_an['synoliko_eur'] = top_an['synoliko_eur'].round(0).astype(int)
    print(f"\n  {dimos}:")
    for name, row in top_an.iterrows():
        print(f"    {name}: {row['anatheseis']} αναθέσεις, €{row['synoliko_eur']:,}")

print("\n" + "="*55)
print("11. ΣΥΓΚΡΙΣΗ ΔΑΠΑΝΩΝ ΠΡΙ/ΜΕΤΑ ΤΟ 2020 ΑΝΑ ΔΗΜΟ")
print("="*55)
print("ΣΗΜΕΙΩΣΗ: χρονικη συσχετιση με Ταμειο Ανακαμψης/ΕΣΠΑ 2021-2027,")
print("οχι αποδεδειγμενη αιτιακη σχεση (η πηγη δεν αναγραφεται στα εγγραφα)")

pre_2020 = df[df['year'] < 2020].groupby('foreas_name')['poso_eur'].sum()
post_2020 = df[df['year'] >= 2020].groupby('foreas_name')['poso_eur'].sum()
compare = pd.DataFrame({'pre_2020': pre_2020, 'post_2020': post_2020}).dropna()
compare = compare[compare['pre_2020'] > 10000]
compare['auxisi_pct'] = ((compare['post_2020'] - compare['pre_2020']) / compare['pre_2020'] * 100).round(0)
top_growth = compare.sort_values('auxisi_pct', ascending=False).head(10).copy()
top_growth['pre_2020'] = top_growth['pre_2020'].round(0).astype(int)
top_growth['post_2020'] = top_growth['post_2020'].round(0).astype(int)
top_growth['auxisi_pct'] = top_growth['auxisi_pct'].astype(int)
display(top_growth)

print("\n--- Εγγραφες 2026 (ατελες ετος, εως Μαιο 2026) ---")
df_2026 = df[df['year'] == 2026]
print(f"Πληθος: {len(df_2026)} εντάλματα | Συνολο: €{df_2026['poso_eur'].sum():,.0f}")

print("\n✅ Αναλυση ολοκληρωθηκε. Τα δεδομενα ειναι ετοιμα για γραφηματα (Datawrapper).")
```
